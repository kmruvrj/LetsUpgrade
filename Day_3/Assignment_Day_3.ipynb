## 1. Write a Program to substract two complex numbers.

```python
#input for two complex numbers
a = int(input("Enter value of a: "))
b = int(input("Enter value of b: "))
c = int(input("Enter value of c: "))
d = int(input("Enter value of d: "))

#code for subtracting two complex numbers
subst_comp_nums = complex(c, d) - complex(a, b) # subracting two complex numbers (c+dj & a+dj)
print (subst_comp_nums)
```

## 2. Write a program to find the fourth root of a number

```python
num = int(input("Please enter the number: "))

fourth_root = num**(1/4)

print(round(fourth_root))
```

## 3. Write a program to swap two numbers with a temporary number

```python
a = int(input("Please enter the First number: "))
b = int(input("Please enter the Second number: "))
temp = a
a = b
b = temp
print("First number after swaping: ", a)
print("Second number after swaping: ", b)
```

## 4. Write a program to swap two numbers without a temporary number

```python
a = int(input("Please enter the First number: "))
b = int(input("Please enter the Second number: "))

a, b = b, a
print("First number after swaping: ", a)
print("Second number after swaping: ", b)
```

## 5. Write a program to convert fahrenheit to Kelvin and Celcius

```python
fahrenheit = float(input("Please enter the Fahrenheit value; "))

celcius = (fahrenheit-32)*(5/9)
print ("The value of celcius is: ", round(celcius, 2),end="deg C\n")
kelvin = celcius + 273.15
print ("The value of Kelvin is: ", round(kelvin, 2), end="K")
```
